```python
import pandas as pd
rows_skip = []
# 读取Excel文件
df_input = pd.read_excel('Copy of xykmx_20231215091635.xlsx',skipfooter=2, header=None, engine='openpyxl')


# 删除包含指定文本的行
df_input = df_input[~df_input[0].str.contains('中国建设银行信用卡交易明细', na=False)]
df_input = df_input[~df_input[0].str.contains('客户名称', na=False)]
df_input = df_input[~df_input[0].str.contains('序号', na=False)]
df_input = df_input[~df_input[7].str.contains('SG', na=False)]
df_input = df_input[~df_input[7].str.contains('约定还款', na=False)]
# 重置索引
df_input.reset_index(drop=True, inplace=True)

df_input
```

```python
columns_exp = ['日期', '类型', '金额', '一级分类', '二级分类', '账户1', '账户2', '备注', '货币', '标签']
columns_input =['序号', '交易日',	'入账日',	'信用卡卡号',	'类型',	'入账币种',	'入账金额',	'交易描述']
```

```python
df_input.columns = columns_input
df_input
```

```python
df_exp = pd.DataFrame(columns=columns_exp)
```

```python
def process_date(date_string_series):
    formatted_date_series = pd.to_datetime(date_string_series, format="%Y%m%d").dt.strftime("%Y年%m月%d日")
    return formatted_date_series
```

```python
transaction_mapping = {"退货退税":"退货"}
category_mapping ={   
'FRESHCO': ['餐饮'],
'APPLE': ['数码', '软件'],
'LUCY':['餐饮', '三餐'],
'SHOPPERS DRUG':['购物'],
'MCDONALD':['餐饮', '三餐'],
'MARSHALLS/HOMES':['购物','日用'],
'MEAL':['餐饮', '三餐'],
'FIDO MOBILE BPY':['通讯'],
'WESTERN DRUG MART':['购物'],
'IVEY BUSINESS SCHOOL': ['餐饮', '三餐'],
'SEPHORA': ['护肤'],
'Temu': ['购物'],
'URBAN PLANET':['购物'],
'COCA COLA': ['餐饮','零食'],
}
```

```python
df_exp['日期'] = process_date(df_input['交易日'])
df_exp['货币'] = 'CNY'
df_exp['金额'] = df_input['入账金额'].abs()
df_exp['备注'] = df_input['交易描述']
df_exp['类型'] = df_input['类型'].map(transaction_mapping)
df_exp['类型'].fillna("支出", inplace=True)
df_exp['账户1'] = '2945'
df_exp['一级分类'] = df_input['交易描述'].apply(lambda x: next((category_mapping[key][0] for key in category_mapping.keys() if key in x), None))
df_exp['二级分类'] = df_input['交易描述'].apply(lambda x: next((category_mapping[key][1] for key in category_mapping.keys() if key in x and len(category_mapping[key]) > 1), None))
```
